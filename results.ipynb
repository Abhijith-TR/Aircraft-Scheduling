```python
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np

def make_smooth_plot(x, y, xlabel, ylabel, title, k_val=9):    
    spl = make_interp_spline(x, y, k=3)
    x_smooth = np.linspace(min(x), max(x), 200)
    y_smooth = spl(x_smooth)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.plot(x_smooth, y_smooth)
    plt.show()

def make_two_smooth_plots(x1, y1, x2, y2, label1, label2, xlabel, ylabel, title):
    spl1 = make_interp_spline(x1, y1, k=5)
    x_smooth1 = np.linspace(min(x1), max(x1), 200)
    y_smooth1 = spl1(x_smooth1)
    spl2 = make_interp_spline(x2, y2, k=5)
    x_smooth2 = np.linspace(min(x2), max(x2), 200)
    y_smooth2 = spl2(x_smooth2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.plot(x_smooth1, y_smooth1, label=label1)
    plt.plot(x_smooth2, y_smooth2, label=label2)
    plt.legend()
    plt.show()
```
